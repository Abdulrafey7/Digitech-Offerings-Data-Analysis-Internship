## Dataset 1 : Group Marks Dataset

```python
df = pd.read_csv("group-marks.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
maths = df['math'].astype(int)
## the astype function doesn't work as it cant convert a string to a number
```

```python
## we dont want to hard code a single case such as "no idea" , what we can do instead is that we can check if there is a character and since we 
## dont want a character in our maths number column we can just put it as NaN

maths = pd.to_numeric(df['math'], errors='coerce')

## here what it does is that it picks up the numbers and convert them to float and for any non numeric values it coerces them (forcefully removes them)
```

```python
maths
```

```python
df['math'] = maths
```

```python
df.info()
```

```python
## now lets fix the other datatypes as well like urdu should be float , same as math and english
## here since urdu is int we can use astype and it wont cause any errors

df['urdu'] = df['urdu'].astype(float)
```

```python
df.info()
```

```python
## now lets take care of the null values from group , math and english
## but first lets analyse the type of data in those columns

print(df['group'].values)
```

```python
## since session is a category lets replace null value with the mode and for math and english we can take the mean value and replace with that

df['math'].fillna(df['math'].mean(), inplace=True)
df['english'].fillna(df['english'].mean(), inplace=True)
df['group'].fillna(df['group'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
original_length = len(df)
original_length
```

```python
df.drop_duplicates(inplace=True)
```

```python
updated_length = len(df)
updated_length
```

```python
df['session'].unique()
```

```python
df['session'] = df['session'].map({'MORNING':'MOR','AFTERNOON':'AFT','EVENING':'EVE','EVE':'EVE','AFT':'AFT','MOR':'MOR'})
```

```python
df['session'].unique()
```

```python
encoded_df = pd.get_dummies(df, columns=['gender', 'group', 'session'], drop_first=False)
encoded_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols_to_scale = ['age', 'scholarship', 'math', 'english','urdu']
encoded_df[cols_to_scale] = scaler.fit_transform(encoded_df[cols_to_scale])
```

```python
encoded_df
```
