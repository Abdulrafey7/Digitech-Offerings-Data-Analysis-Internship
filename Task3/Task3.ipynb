```python
## null values ko mean sa replace kr diya
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
print(df['Age'])
```

```python
## embarked kioun keh categorical feature hy toh iska mean ni la skta toh mode la liya jo most frequent value sa replace 
## kr raha hy
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
## kioun keh cabin ma bohot ziyada null values tha aur cabin number itna usefull ni hy toh usko drop krdiya
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.head()
```

```python
len(df)
```

```python
df.drop_duplicates(inplace=True)
```

```python
len(df)
## no difference in lenght after removing duplicates matlab 0 duplicates
```

```python
## age ko int krdiya from float because we want a fixed age 
df['Age'] = df['Age'].astype(int)
```

```python
df.rename(columns={'Pclass': 'PassengerClass', 
                   'SibSp': 'SiblingsSpouses', 
                   'Parch': 'ParentsChildren'}, inplace=True)
```

```python
# Apply ko use krka aik new column banaya if a person is above or equal 18 they are a adult
df['Adult'] = df['Age'].apply(lambda x: True if x >= 18 else False)
```

```python
# Another example: Convert 'Sex' column to numeric (0 for male, 1 for female)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
print("\n----- Final Dataset Summary -----")
print(f"Number of Rows: {df.shape[0]}")
print(f"Number of Columns: {df.shape[1]}")
print(f"Duplicate Rows Removed: 0")
print(f"Missing Values After Handling:\n{df.isnull().sum()}")

print("\n----- Final Dataset Sample -----")
print(df.head())
```
