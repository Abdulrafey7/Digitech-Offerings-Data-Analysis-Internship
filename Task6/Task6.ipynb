```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("spotify_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["album_release_date"] = pd.to_datetime(df["album_release_date"])
```

```python
df.info()
```

```python
df.head()
```

### Bar Plot
This shows the popularity of 10 most popular tracks , as we can see the mean is around 52 and the most popular tracks in the 90's

```python
df["track_popularity"].mean()
```

```python
top_tracks = df.nlargest(10, "track_popularity")

plt.figure(figsize=(12,6))
plt.bar(top_tracks["track_name"], top_tracks["track_popularity"])
plt.xticks(rotation=75)
plt.xlabel("Track Name")
plt.ylabel("Popularity")
plt.title("Top 10 Most Popular Tracks")
plt.show()
```

### Line Plot 
This shows us how a artist's followers affect thier popularity

```python
df_sorted = df.sort_values("artist_followers")

plt.figure(figsize=(10,6))
plt.plot(df_sorted["artist_followers"], df_sorted["artist_popularity"], linestyle='-', linewidth=2)
plt.xlabel("Artist Followers")
plt.ylabel("Artist Popularity")
plt.title("Artist Popularity vs Artist Followers")
plt.tight_layout()
plt.show()
```

### Pie Plot
Shows us how album types are distributed

```python
plt.figure(figsize=(6,6))
df["album_type"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Album Type Distribution")
plt.ylabel("")
plt.show()
```

### Coorelation Heatman
This shows us how different attributes have affect on eachother like , the higher the value the more they have on eachother like how artist popularity and track popularity have strong influence like if a artist is popular their track will also be popular

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(numeric_df[["track_popularity", "artist_popularity", "artist_followers", "track_duration_min"]])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="album_type", order=df["album_type"].value_counts().index)
plt.title("Count of Album Types")
plt.xlabel("Album Type")
plt.ylabel("Count")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14,5))

sns.histplot(df["track_popularity"], ax=ax[0], kde=True)
ax[0].set_title("Track Popularity Distribution")

sns.boxplot(data=df, x="track_duration_min", ax=ax[1])
ax[1].set_title("Track Duration Boxplot")

plt.tight_layout()
plt.show()
```
